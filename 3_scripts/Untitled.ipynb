```python
# creating the dictionaries

dict2 = {'Real_teacher':['Ms Chua','Chia','Ms Noone','Ms Yu Qian','Mr Chen','Mr Ng Say Tiong','H. Halfmouw','Ms Hogan','Mrs O Leary','S. H√∂lscher', 'Ms Lee Wing','Clarke','Duffy','Lang Fui Ying']}

# converting to pandas dataframes
#dframe1['Teacher'] = Teacher_name['Teacher']
dframe2 = pd.DataFrame(dict2)

# empty lists for storing the matches
# later
mat1 = []
mat2 = []
p = []

# printing the pandas dataframes
print("First dataframe:\n", Teacher_name,
	"\nSecond dataframe:\n", dframe2)

# converting dataframe column
# to list of elements
# to do fuzzy matching
list1 = Teacher_name['Teacher'].tolist()
list2 = dframe2['Real_teacher'].tolist()
```

```python
# taking the threshold as 80
threshold = 70

# iterating through list1 to extract
# it's closest match from list2
for i in list1:
	mat1.append(process.extractOne(
	i, list2, scorer=fuzz.token_set_ratio))
Teacher_name['matches'] = mat1

# iterating through the closest matches
# to filter out the maximum closest match
for j in Teacher_name['matches']:
	if j[1] >= threshold:
		p.append(j[0])
	mat2.append(",".join(p))
	p = []


# storing the resultant matches back
# to dframe1
Teacher_name['matches'] = mat2
print("\nDataFrame after Fuzzy matching using token_set_ratio():")
pd.set_option('display.max_rows', Teacher_name.shape[0]-30)
Teacher_name
```
